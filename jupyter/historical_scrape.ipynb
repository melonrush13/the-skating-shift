```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from textblob import TextBlob
```

```python
# Open 2019 Tweets
tweets2019 = "..//scraped_tweets/2019_snscrape_tweets.csv"
scrapedTweets2019 = pd.read_csv(tweets2019, lineterminator='\n')
scrapedTweets2019.info()
```

```python
#2019
# Get Sum of RollerSkate Tweets Per Month
df = pd.DataFrame()
df['datetime'] = scrapedTweets2019['datetime']
# Covert Pandas core series to type datetime 
df['datetime']= pd.to_datetime(df['datetime'])
# Convert datetime using dt to year
df['datetime']= df['datetime'].dt.month
# add new col to count for tweet per row
df['tweetcount'] = 1

rollerskatingAgg2019 = pd.pivot_table(data=df, index='datetime',
               values='tweetcount',
               aggfunc='sum').reset_index().rename(columns={'tweetcount':'tweetsperday'})

print(rollerskatingAgg2019)
rollerskatingAgg2019.to_csv("rsSum19.csv")
```

```python
# Plot Tweets from 2016 to 2020
x20 = rollerskatingAgg2020['datetime']
y20 = rollerskatingAgg2020['tweetsperday']
x19 = rollerskatingAgg2019['datetime']
y19 = rollerskatingAgg2019['tweetsperday']
x18 = rollerskatingAgg2018['datetime']
y18 = rollerskatingAgg2018['tweetsperday']
x17 = rollerskatingAgg2017['datetime']
y17 = rollerskatingAgg2017['tweetsperday']
x16 = rollerskatingAgg2016['datetime']
y16 = rollerskatingAgg2016['tweetsperday']

fig, axs = plt.subplots(1,1)
axs.plot(x20,y20, label = '2020')
axs.plot(x19,y19, label = '2019')
axs.plot(x18,y18, label = '2018')
axs.plot(x17,y17, label = '2017')
axs.plot(x16,y16, label = '2016')
axs.set_xlabel('Months')
axs.set_ylabel('Num Tweets')
plt.legend()
axs.set_title("Rollerskating Tweets 2016-2020")

plt.show()
```
